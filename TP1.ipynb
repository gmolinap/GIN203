Question 1. Tracez la courbe représentant la taille de fenêtre de congestion de l'émetteur en fonction du temps, 
utilisez pour cela le fichier f-out.cls obtenu par le traitement du script fenetre-tcp2.tcl par NS. Remarquez que ce script 
considère un cas sans perte. Interprétez cette courbe. 

```python
import matplotlib.pyplot as plt

# Datos proporcionados
data = open('f-out.txt').read()

# Dividir los datos en listas de tiempo (x) y tamaño de la ventana (y)
lines = data.strip().split('\n')
x, y = zip(*[map(float, line.split()) for line in lines])

# Crear el gráfico
plt.plot(x, y, label='Tamaño de la ventana de congestión')
plt.xlabel('Tiempo')
plt.ylabel('Tamaño de la ventana de congestión')
plt.title('Tamaño de la ventana de congestión en función del tiempo')
plt.legend()
plt.show()
```

El tamaño de la ventana de congestion tiene un crecimiento de 0.1 a 0.4 y se estabiliza en el tiempo 0.5.

Question 3. Simulez ce réseau  à l'aide du fichier busTCP.tcl. Dès lors, vous obtiendrez le fichier p-out.tcl contenant le nombre de paquets reçus pour les deux connexions TCP. 
Tracez et interprétez la courbe du débit instantané des deux connexions TCP. Que pourrait-on en conclure sur le partage des ressources entre deux flux TCP  ? 


À la date t = 0 s, une connexion TCP est ouverte entre A et B pour transporter un flux FTP continu. 

À la date t = 1 s, une connexion TCP est ouverte entre A' et B' pour transporter un flux FTP qui stoppera avant la fin de simulation (la durée totale de simulation est de 4 s). 


```python
import matplotlib.pyplot as plt
import numpy as np

# Datos proporcionados
data = open('p-out.txt').read()

# Dividir los datos en listas de tiempo (x), y1, y2, y3, y4 (tamaño de la ventana de congestión)
lines = data.strip().split('\n')
data_list = [list(map(float, line.split())) for line in lines]
time, y1, y2, y3, y4 = zip(*data_list)

# Crear el gráfico
plt.figure(figsize=(10, 6))
plt.plot(time, y1, label='y1')
plt.plot(time, y2, label='y2')
plt.plot(time, y3, label='y3')
plt.plot(time, y4, label='y4')

plt.xlabel('Tiempo')
plt.ylabel('Tamaño de la ventana de congestión')
plt.title('Tamaño de la ventana de congestión en función del tiempo')
plt.legend()
plt.show()
```
